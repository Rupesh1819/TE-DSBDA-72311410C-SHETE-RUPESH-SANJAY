#### Dataset Description
The Titanic dataset contains information about passengers aboard the RMS Titanic sinking, including features like age, gender, ticket class, and fare. It is commonly used for classification tasks to predict whether a passenger survived (1) or not (0). The dataset requires preprocessing due to missing values and categorical data.

```python
#import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#load Dataset
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#check missing values
df.isnull().sum()
```

```python
df = df[['Age','Sex','Survived']]
df = df.dropna(subset=['Age'])
df.head()
```

```python
df.isnull().sum()
```

```python
#Box plot 
plt.figure(figsize=(10,6))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=df)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by Gender and Survival')
```

## Observations
- Females have higher survival rate.
- Younger passengers survived more.
- Males have more non-survivors.
- Outliers show elderly passengers.
