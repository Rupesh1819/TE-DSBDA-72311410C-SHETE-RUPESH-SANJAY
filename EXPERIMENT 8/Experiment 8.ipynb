#### Dataset Description
The Titanic dataset contains information about passengers aboard the RMS Titanic sinking, including features like age, gender, ticket class, and fare. It is commonly used for classification tasks to predict whether a passenger survived (1) or not (0). The dataset requires preprocessing due to missing values and categorical data.

```python
#Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load Dataset 
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
#Dataset Description 
df.head()
```

```python
df.describe()
```

```python
#Data Visualization - Pattern Finding

#Count plot (Survival count)
sns.countplot(x='Survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
```

```python
# Gender vs Survival
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival based on Gender")
```

```python
# Passenger Class vs Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival based on Passenger Class")
```

```python
# Age distribution
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
```

```python
#Fare Distribution (Histogram)
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare in $")
plt.ylabel("Frequency")
```
