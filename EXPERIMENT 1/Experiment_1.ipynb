**Type converting**

here Pclass is not in correct type so change it into category

```python
df['Pclass'] = df['Pclass'].astype('category')
```

## Data Normalization
Normalization scales numeric data into a fixed range (commonly 0–1)

```python
df['Age_Normalized'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df[['Age', 'Age_Normalized']].head()
```

## Convert Categorical Variables into Quantitative Variables
Machine learning models require numeric values.
We convert categorical variables using **One‑Hot Encoding**.

```python
df_encoded = pd.get_dummies(df, columns=['Sex', 'Embarked'])
```
