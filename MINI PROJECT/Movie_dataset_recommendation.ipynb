```python
MULTI_WORD_GENRES = {
    'Science Fiction': 'ScienceFiction',
    'TV Movie': 'TVMovie',
}

def convert(obj):
    if not isinstance(obj, str):
        return []
    for phrase, replacement in MULTI_WORD_GENRES.items():
        obj = obj.replace(phrase, replacement)
    return obj.split()
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# cast is a plain space-separated string of names, group pairs of words into full names
def convert2(obj):
    if not isinstance(obj, str):
        return []
    words = obj.split()
    names = []
    i = 0
    while i < len(words) and len(names) < 3:
        if i + 1 < len(words):
            names.append(words[i] + words[i + 1])
            i += 2
        else:
            names.append(words[i])
            i += 1
    return names
```

```python
movies['cast'].apply(convert2)
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head(1)
```

```python
# director column is already a plain text string in movie_dataset.csv
# wrap it into a list so it merges cleanly with other list columns
def fetch_director(obj):
    if isinstance(obj, str) and obj.strip():
        return [obj.strip()]
    return []
```

```python
movies['director'].apply(fetch_director)
```

```python
movies['crew'] = movies['director'].apply(fetch_director)
```

```python
movies.head()
```

```python
#convert the overview into the list
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else [])
```

```python
movies.head()
```

```python
#remove the space between the word
movies['overview'] = movies['overview'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
movies['genres'] = movies['genres'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
movies['cast'] = movies['cast'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
movies['crew'] = movies['crew'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('danced')
```

```python
# map the vector and find the distance between them 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python

def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
