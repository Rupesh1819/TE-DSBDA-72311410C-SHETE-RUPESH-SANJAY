```python
#import required Libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Download required resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
#Sample Document 
document = """
Text analytics is the process of deriving meaningful information from text. 
It involves preprocessing steps like tokenization, stop word removal, stemming, and lemmatization.
"""
```

```python
print("Original Document:\n", document)
```

```python
#Tokenization 
Tokens =word_tokenize(document)
print("\nTokens:\n", Tokens)
```

```python
#POS tagging
pos_tags=pos_tag(Tokens)
print("\nPOS Tags:\n", pos_tags)
```

```python
#Stop word removals 
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in Tokens if word.lower() not in stop_words]
print("\nAfter Stop Words Removal:\n", filtered_tokens)
```

```python
#Stemming 
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_tokens]
print("\nStemmed Words:\n", stemmed_words)
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("\nLemmatized Words:\n", lemmatized_words)
```

```python
#TF-IDF Representation
documents = [
    "Text analytics extracts useful insights from text data",
    "Natural language processing is part of text analytics",
    "TF IDF helps in understanding importance of words"
]
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
print("\nTF-IDF Feature Names:\n", vectorizer.get_feature_names_out())
```

```python
tfidf_matrix = vectorizer.fit_transform(documents)
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
```
